```python
from tkinter import *
root=Tk()
root.title('Calculo de Presion Hidrostatica')
root.geometry('500x500')
root.grid_rowconfigure(0, weight=1)
root.grid_rowconfigure(1, weight=1)
root.grid_columnconfigure(0, weight=1)
root.resizable(False, False)

# Calculos
def getValues():
    dval=str(dinput.get())
    h1val=str(h1input.get())
    h2val=str(h2input.get())
    Values = "d= "+dval+" h1="+h1val+" h2="+h2val
    ValuesLabel =Label(frame2a,text=Values)
    ValuesLabel.grid(row=3, column=1)
    global dcalc
    global h1calc
    global h2calc
    dcalc = int(dval)
    h1calc = float(h1val)
    h2calc = float(h2val)
    p = dcalc*(9.82)*(h2calc-h1calc)
    calcLabel =Label(frame2a,text="Presion (p) = "+str(round(p,2))+" Pa")
    calcLabel.grid(row=2, column=1)
    w.delete("deltaH")
    w.create_line(100, 200*(int(h1input.get()/3)), 100, 200*(h2input.get()/3), fill="red",width=3,tags = "deltaH")
    

#Frames
frame1 =Frame(root,bg="lightblue")
frame2 =Frame(root,bg="lavender",width=500,height=250)
frame2a=Frame(frame2,bg="lavender", width=250, height=250)
frame2b=Frame(frame2,bg="lavender", width=250, height=250)
frame1.grid(row=0, sticky=S+N+E+W)
frame2.grid(row=1, sticky=S+N+E+W)
frame2a.grid(row=1,column=1,sticky=S+N+E+W)
frame2b.grid(row=1,column=2,sticky=S+N+E+W)

# Mostrar en Pantalla

frame1.grid_rowconfigure(1, weight=1)
frame1.grid_rowconfigure(2, weight=1)
frame1.grid_rowconfigure(3, weight=1)
frame1.grid_columnconfigure(0, weight=1)
frame1.grid_columnconfigure(1, weight=1)

frame2.grid_rowconfigure(1, weight=1)
frame2.grid_columnconfigure(1, weight=0)
frame2.grid_columnconfigure(2, weight=0)

frame2a.grid_rowconfigure(1,weight=1)
frame2a.grid_rowconfigure(2,weight=1)
frame2a.grid_rowconfigure(3,weight=1)
frame2a.grid_columnconfigure(1, weight=1)

frame2b.grid_rowconfigure(1,weight=1)
frame2.grid_columnconfigure(1, weight=1)

# Inputs
dlabel=Label(frame1,text="Ingresa el valor de la densidad(d):",bg="lightblue")
dinput =Scale(frame1, from_=0, to=1500, orient=HORIZONTAL,length=200,bg="lightblue")
h1label=Label(frame1,text="Ingresa el valor de h1:",bg="lightblue")
h1input =Scale(frame1, from_=0, to=3, orient=HORIZONTAL,length=200,digits = 2, resolution = 0.1,bg="lightblue")
h2label=Label(frame1,text="Ingresa el valor de h2:",bg="lightblue")
h2input =Scale(frame1, from_=0, to=3, orient=HORIZONTAL,length=200,digits = 2, resolution = 0.1,bg="lightblue")

dlabel.grid(row=1, column=0)
dinput.grid(row=1, column=1)
h1label.grid(row=2, column=0)
h1input.grid(row=2, column=1)
h2label.grid(row=3, column=0)
h2input.grid(row=3, column=1)

#Boton
pbtn=Button(frame2a,text="Calcular",command=getValues,bg="lightblue")
pbtn.grid(row=1, column=1)
calcLabel =Label(frame2a,text="")
calcLabel.grid(row=2, column=1)
ValuesLabel =Label(frame2a,text="")
ValuesLabel.grid(row=3, column=1)


#Tanque
w = Canvas(frame2b, width=200, height=200,bg="lavender")
w.create_oval(0, 10, 200, 20, fill="#476042",width=3)
w.create_line(3, 15, 3, 200, fill="#476042", width=3)
w.create_line(199, 15, 199, 200, fill="#476042", width=3)
w.create_line(0,200, 200, 200, fill="#476042", width=3)
w.create_rectangle(5,25, 198, 200, fill="blue")
w.grid(row=1)


#Fin del programa
root.mainloop()
```
